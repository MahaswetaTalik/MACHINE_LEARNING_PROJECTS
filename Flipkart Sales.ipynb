# Adding dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing our training data
train_data = pd.read_csv('C:/Users/KIIT0001/Desktop/MAHASWETA/MY PROJECTS/DATASETS/train.csv')
train_data.head(3)
```

```python
# importing our test data
test_data = pd.read_csv('C:/Users/KIIT0001/Desktop/MAHASWETA/MY PROJECTS/DATASETS/test.csv')
test_data.head(3)
```

```python
sales_data = pd.concat([train_data, test_data], axis=0)  # axis=0 merges them row-wise

# Reset index to ensure continuity
sales_data.reset_index(drop=True, inplace=True)

# View the merged dataset
print(sales_data.head())
```

# Handling missing values

```python
# counting missing values in each column
sales_data.isnull().sum()
```

```python
# handling missing values
for column in sales_data.columns:
    if sales_data[column].dtype == 'float64':  # Numeric (float)
        sales_data[column].fillna(sales_data[column].mean(), inplace=True)
    elif sales_data[column].dtype == 'int64':  # Numeric (int)
        sales_data[column].fillna(sales_data[column].mean(), inplace=True)  # Using mean for consistency
    elif sales_data[column].dtype == 'object':  # Categorical (string)
        sales_data[column].fillna(sales_data[column].mode()[0], inplace=True)  # Mode for categorical
        
sales_data.isnull().sum()
```

# Data visualization

```python
# Sales Performance Across Platform

# Group by platform and average price
platform_sales = sales_data.groupby("platform")["price1"].mean().reset_index()

plt.figure(figsize=(8, 4))
sns.barplot(data=platform_sales, x="platform", y="price1", palette="dark",legend=False)
plt.title("Average Sales Price by Platform")
plt.xlabel("Platform")
plt.ylabel("Average Price")
plt.xticks(rotation=45)
plt.show()
```

```python
# Customer Ratings Distribution

plt.figure(figsize=(6, 2))
sns.histplot(sales_data["Rating"].dropna(), bins=20, kde=True, color="skyblue")
plt.title("Distribution of Customer Ratings")
plt.xlabel("Ratings")
plt.ylabel("Frequency")
plt.show()
```

```python
# Discount vs. Actual Price

plt.figure(figsize=(8, 4))
sns.scatterplot(data=sales_data, x="actprice1", y="Offer %", alpha=0.6)
plt.title("Discount vs. Actual Price")
plt.xlabel("Actual Price")
plt.ylabel("Offer Percentage")
plt.show()
```

```python
# Review Counts vs. Star Rating

# Selecting relevant columns
corr_data = sales_data[["noreviews1", "star_5f", "star_4f", "star_3f", "star_2f", "star_1f"]].corr()

plt.figure(figsize=(5, 3))
sns.heatmap(corr_data, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Reviews and Star Ratings")
plt.show()
```

# EDA

```python
# Distribution & Outliers in Key Features

print(sales_data["price1"].describe())  # Price distribution
print(sales_data["Rating"].describe())  # Rating summary

# Checking outliers using interquartile range (IQR)
Q1 = sales_data["price1"].quantile(0.25)
Q3 = sales_data["price1"].quantile(0.75)
IQR = Q3 - Q1
outliers = sales_data[(sales_data["price1"] < Q1 - 1.5 * IQR) | (sales_data["price1"] > Q3 + 1.5 * IQR)]
print(f"Number of outliers in price: {outliers.shape[0]}")
```
